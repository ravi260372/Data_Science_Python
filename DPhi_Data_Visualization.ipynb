# Scatter Plot  
* A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for different numeric variables.  
* The position of each dot on the horizontal and vertical axis indicates values for an individual data point.  
* Scatter plots are used to observe relationships between variables.  
* To create a scatter plot in Matplotlib we can use the .scatter() method  
* A scatter plot can be useful for identifying other patterns in data.  
    We can divide data points into groups based on how closely sets of points cluster together.    
    Scatter plots can also show if there are any unexpected gaps in the data and if there are any outlier points. (Look at the 2 points away from rest of the data in the scatter plot. Those are outliers.)  
This can be useful if we want to segment the data into different parts, like categorising users into different groups.


```
# scatter plot crime_rate v/s % senior
plt.scatter(df.crime_rate , df.percent_senior) # plotting the sactter plot
plt.show(); # visualize the plot
```

# Histogram  
A histogram is a graphical display of data using bars(rectangles) of different heights.

***Parts of a Histogram:***
1. The title: The title describes the information included in the histogram.
2. X-axis: The X-axis are intervals that show the scale of values which the measurements fall under. These intervals are also called bins.
3. Y-axis: The Y-axis shows the number of times that the values occurred(frequency) for each interval on the X-axis.
4. The bars: The height of the bar shows the number of times that the values occurred within the interval, while the width of the bar shows the interval that is covered.

***Create a Histogram***
Matplotlib can be used to create histograms using the hist() method. 

**Parameters:**
* x(n,) : this takes either a single array or a sequence of arrays which are not required to be of the same length.  
* bins : intervals of any quantity  
If the bins are: [1, 2, 3, 4]  
then the first bin is [1, 2) (including 1, but excluding 2) and the second [2, 3). The last bin, however, is [3, 4], which includes 4.  

**Applications:**  
Histograms are a very common type of plots when we are looking at numerical-continuous data like height , weight, stock prices, waiting time for a customer, etc.

Histograms help us visualize the general distributional features of a dataset variables. we can visualize where the peaks of the distribution are, whether the distribution is skewed or symmetric, and if there are any outliers.


```
# plot a histogram of % senior citizens
plt.hist(df.percent_senior)
plt.title('Distribution of Senior Citizens')
plt.show();
```

* The x-axis is the "number of values" in each bin i.e. "frequency"  
* The y-axis is the bins created (default or user defined).

# Bar Chart  
Bar charts are one of the most common types of graphs that are used to visualize data associated with the categorical variables (counts).  

**Parts of a Bar Chart:**  

The title: The title describes the information included in the histogram.
X-axis: The X-axis are categories / levels of a categorical varible.
Y-axis: The Y-axis shows the number of times that the values occurred(frequency) for each category on the X-axis.
The bars: The height of the bar shows the number of times that the values occurred within the interval  

**Create a Bar Chart**  
Pyplot provides a bar() method to make bar graphs which take the following arguments: 
* categorical variables  
* values of the categorical variable 
* color (if you want to specify any). 

Applications:
Bar graphs are used to compare observations between different groups or to trace changes over time.



```
# Verical distribution of crime-rate by region
plt.bar(df.region , df.crime_rate)
plt.title('Crime Rate across Regions')
plt.show();
```

```
# Horizontal distribution of crime-rate by region
plt.barh(df.region , df.crime_rate)
plt.title('Crime Rate across Regions')
plt.show();
```

# Bar Charts with multiple quantities
When comparing several quantities and when changing one variable, we might want a bar chart where we have bars of one color for one quantity value.

We can plot multiple bar charts by playing with the thickness and the positions of the bars.

```
divisions = ["A", "B", "C", "D", "E"]
division_avg = [70, 82, 73, 65, 68]
boys_avg = [68, 67, 77, 61, 70]

# Using the NumPy arange function to generate values for index between 0-4.
# Here,stop is 5, start is 0, and step is 1
index = np.arange(5)
width = 0.30

plt.bar(index, division_avg, width, color="green", label="Division Marks")
plt.bar(index+width, boys_avg, width, color="blue", label="Boys Marks")

plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

# Stacked Bar Chart
The stacked bar chart stacks bars that represent different groups on top of each other. The height of the resulting bar shows the combined result of the groups.

The optional bottom parameter of the pyplot.bar() function allows you to specify a starting value for a bar. Instead of running from zero to a value, it will go from the bottom to the value. The first call to pyplot.bar() plots the blue bars. The second call to pyplot.bar() plots the green bars, with the bottom of the blue bars being at the top of the green bars.

```
divisions = ["A", "B", "C", "D", "E"]
girls_avg = [72, 97, 69, 69, 66]
boys_avg = [68, 67, 77, 61, 70]

index = np.arange(5)
width = 0.50

plt.bar(index, boys_avg, width, color="green", label="Boys Marks")
plt.bar(index, girls_avg, width, color="blue", label="Girls Marks", bottom=boys_avg)

plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

# Advanced Plots : Box Plot 

With a boxplot, we can extract the same insights as with an histogram. And while we can visualize the shape of the distribution with an histogram, a boxplot highlights the summary metrics that give the distribution its shape. The summary metrics we can extract from a boxplot are:

* Quantiles, specifically the first and third quantiles, which correspond to the 25th and 75th percentiles.  
* Median, the mid-point in the distribution, which also corresponds to the 50th percentile.  
* Interquartile range (IQR), the width between the third and first quantiles. Expressed mathematically, we have IQR = Q3 — Q1.  
* Min, minimum value in the dataset excluding outliers, which corresponds to Q1–1.5xIQR.  
* Max, maximum value in the dataset, excluding outliers, which corresponds to Q3+ 1.5xIQR.  

```
# Imports
import pandas as pd
import numpy as np # for calculating standard deviation and mean
import matplotlib.pyplot as plt # for improving our visualizations
from sklearn import datasets
import seaborn as sns

# to remove warnings
import warnings
warnings.filterwarnings('ignore')
```

```
# Load Iris dataset
iris = datasets.load_iris()
```

```
# Preparing Iris dataset
iris_data = pd.DataFrame(data=iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
iris_target = pd.DataFrame(data=iris.target, columns=['species'])
iris_df = pd.concat([iris_data, iris_target], axis=1)
```

```
# Add species name
iris_df['species_name'] = np.where(iris_df['species'] == 0, 'Setosa', None)
iris_df['species_name'] = np.where(iris_df['species'] == 1, 'Versicolor', iris_df['species_name'])
iris_df['species_name'] = np.where(iris_df['species'] == 2, 'Virginica', iris_df['species_name'])
iris_df.sample(5)
```

```
# Prepare petal length by species datasets
setosa_petal_length = iris_df[iris_df['species_name'] == 'Setosa']['petal_length']
versicolor_petal_length = iris_df[iris_df['species_name'] == 'Versicolor']['petal_length']
virginica_petal_length = iris_df[iris_df['species_name'] == 'Virginica']['petal_length']
```

```
# Visualize petal length distribution for all species
fig, ax = plt.subplots(figsize=(12, 7))

# Add major gridlines in the y-axis
ax.grid(color='grey', axis='y', linestyle='-', linewidth=0.25, alpha=0.5)

# Set plot title
ax.set_title('Distribution of petal length by species')

# Set species names as labels for the boxplot
dataset = [setosa_petal_length, versicolor_petal_length, virginica_petal_length]
labels = iris_df['species_name'].unique()

ax.boxplot(dataset, labels=labels)

plt.show();
```

We can extract a few insights from this plot:  
* Iris Setosa has a much smaller petal length than Iris Versicolor and Virginica. It ranges from approximately 1 to 2 centimeters.
* The range of petal length of Iris Virginica is bigger than both the ranges of values for Iris Setosa and Versicolor. We can see that from how tall the box is for Iris Virginica compared to the other two.
* Both Iris Setosa and Veriscolor have outliers.

# Advanced Plots : Pair Pplot  

While there are an almost overwhelming number of methods to use in EDA, one of the most effective starting tools is the pair plot (also called a scatterplot matrix). A pairs plot allows us to see both distribution of single variables and relationships between two variables. Pair plots are a great method to identify trends for follow-up analysis.The default pairs plot in seaborn only plots numerical columns.  

The pairs plot builds on two basic figures, the histogram and the scatter plot. The histogram on the diagonal allows us to see the distribution of a single variable while the scatter plots on the upper and lower triangles show the relationship (or lack thereof) between two variables. For example, the left-most plot in the third row shows the scatter plot of sepal_length versus petal_length.

The default pairs plot by itself often gives us valuable insights. We can see that petal lengths are positively correlated with petal width.

While this plot alone can be useful in an analysis, we can find make it more valuable by coloring the figures based on a categorical variable such as species.

```
# basic pair plot of iris features
sns.pairplot(iris_df);
```

```
# pair plot by species
sns.pairplot(iris_df, hue='species');
```

# Advanced Plots : Density - KDE Plots
A density plot is a smoothed, continuous version of a histogram estimated from the data. The most common form of estimation is known as kernel density estimation. In this method, a continuous curve (the kernel) is drawn at every individual data point and all of these curves are then added together to make a single smooth density estimation. The kernel most often used is a Gaussian (which produces a Gaussian bell curve at each data point).

Density Plots in Seaborn
To make density plots in seaborn, we can use either the distplot or kdeplot function. For example, we can make a density plot showing all sepal lengths on top of the corresponding histogram:

```
# Density Plot and Histogram of Sepal Length
sns.distplot(iris_df['sepal_length'], hist=True, kde=True,
bins=int(180/5), color = 'darkblue',
hist_kws={'edgecolor':'black'},
kde_kws={'linewidth': 4})
plt.show();
```

```
# Density Plot of Sepal Length by Species
sns.kdeplot(iris_df['sepal_length'], hue=iris_df['species'],shade=True) # for different species
plt.show();
```

```
# Density Plot of Petal Length by Species
sns.kdeplot(iris_df['petal_length'], hue=iris_df['species'],shade=True) # for different species
plt.show();
```

# Advanced Plots : Bubble Plot
Bubble plots are an improved version of the scatter plot. In a scatter plot, there are two dimensions x, and y. In a bubble plot, there are three dimensions x, y, and z. Where the third dimension z denotes weight. Bubble plots give more information visually than a two dimensional scatter plot.To make bubble plot in Seaborn, we are able to use scatterplot() function in Seaborn with a variable specifying size argument in addition to x and y-axis variables for scatter plot.

In this bubble plot instance, we have length = 'sepal_length'. And this will create a bubble plot with unique bubble sizes based at the sepal length variable.With the argument “sizes” in Seaborn’s scatterplot() function, we can specify ranges for the bubble sizes. In this bubble plot example below, we used sizes=(20,500)

```
# create the scatter plot of 'petal length v/s petal width' with "sepal length as size"
plt.figure(figsize=(10, 6))
sns.scatterplot(x="petal_length",
y="petal_width",
size="sepal_length",
sizes=(20,500),
alpha=0.5,
data=iris_df)

# Put the legend out of the figure
plt.legend(title = 'Sepal Length', bbox_to_anchor=(1.01, 1),borderaxespad=0)
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.tight_layout()

plt.show();
```

In the above bubble plots, the smallest bubbles relates to lowest weight and the larger bubbles relates to biggest weight. The legend on the top right corner tells the sepal length corresponding to the bubble size. Larger the bubble size, greater the sepal length for the particular observation.

# Advanced Plots : Joint Plot  
Joint Plot draws a plot of two variables with bivariate and univariate graphs. It uses the Scatter Plot and Histogram. Joint Plot can also display data using Kernel Density Estimate (KDE) and Hexagons. We can also draw a Regression Line in Scatter Plot.  

We can observe in the plot below that the density of points is more for sepal length around 5.5 to 6.5 and sepal width around 2.5 to 3.5

```
sns.set_style("dark")
sns.jointplot(x='sepal_length', y='sepal_width',data=iris_df);
```

The plot below is a modified version of the earlier jointplot. This plot contains a regression line (i.e. a straight line) passing through the points such that the sum os square of distances from each point to the straight line is minimum possible. Both the histogram contains a distribution line over the histogram. This is the density plot of both the variables - sepal_length and sepal_width.

```
sns.set_style("dark")

# Add regression line to scatter plot and kernel density estimate to histogram
sns.jointplot(x='sepal_length', y='sepal_width',data=iris_df, kind = 'reg')
```
